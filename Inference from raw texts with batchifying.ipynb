```python
for batch_size in [32, 64, 128, 256, 512]:
    total_time = 0
    start = time.time()
    segmented_texts = uitws_v1.segment(texts=texts * 10**3, pre_tokenized=False, batch_size=batch_size)
    end = time.time()
    total_time += (end - start)
    nb_sentences = sum([len(paragraph) for doc in segmented_texts for paragraph in doc])
    print("batch_size:", batch_size)
    print("speed of word segmentation (seconds/sentence):", total_time/nb_sentences)
    print("speed of word segmentation (sentences/second):", nb_sentences/total_time)
    print("")
```

```python
print("number of sentence in this experiment:", nb_sentences)
```

```python
length_of_sentences = [len(sentence) for doc in segmented_texts for paragraph in doc for sentence in paragraph]
print("average_length in this experiment:", sum(length_of_sentences)/len(length_of_sentences))
```
